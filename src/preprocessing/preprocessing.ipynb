```python
import pandas as pd
import re
from tqdm import tqdm;  tqdm.pandas()
```

```python
def load_data(path):
    """
    주어진 경로에서 데이터를 로드합니다.
    
    Args:
        path (str): 데이터 파일 경로
    
    Returns:
        pd.DataFrame: 로드된 데이터프레임
    """
    df = pd.read_csv(path, encoding='utf-8')
    print("Data loaded successfully")
    return df


def remove_special_chars(text):
    """
    문자열에서 숫자, 영문, 한글, 공백을 제외한 특수문자를 제거합니다.
    """
    return re.sub(r"[^0-9a-zA-Zㄱ-ㅎㅏ-ㅣ가-힣 ]", '', str(text))
```
